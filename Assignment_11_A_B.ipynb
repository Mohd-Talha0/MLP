```python
data = pd.read_csv("/content/AirQualityUCI.csv")

data.shape
```

```python
data.head()
```

```python

data.drop(columns=['Date', 'Time', 'Unnamed: 15', 'Unnamed: 16'], inplace=True)
```

```python
data.head()
```

```python
data2 = data.copy()
```

```python
data2.head()
```

```python
data2.dropna(axis=0, inplace=True)
```

```python
data2.head()
```

```python
data2.shape
```

```python
correlation_coefficient = data2['T'].corr(data['RH'])

print("Correlation coefficient between 'education-num' and 'capital-gain':", correlation_coefficient)
```

```python
selector = lambda df: df['AH'] > 1.5
data2_cut = data2.loc[selector]
data2_cut
```

```python
y = data2['AH']

X = data2.drop(['AH'], axis = 1 )
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

# Create a pipeline with StandardScaler and MLPRegressor
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # StandardScaler for scaling features
    ('regressor', MLPRegressor(hidden_layer_sizes=(50, 50, 50),  # Three hidden layers with 50 neurons each
                                tol=1e-2,                       # Tolerance for optimization
                                alpha=1e-4,                     # L2 penalty (regularization term) parameter
                                solver='adam',                  # Optimization algorithm
                                learning_rate_init=0.1,         # Initial learning rate
                                max_iter=50,                    # Maximum number of iterations
                                random_state=1))                # Random state for reproducibility
])

# You can access the individual components of the pipeline using named_steps attribute
# For example, to access the scaler component:
# scaler = pipeline.named_steps['scaler']

# To access the regressor component:
# regressor = pipeline.named_steps['regressor']

# Now you can use this pipeline object for fitting your data and making predictions.

# Fit the pipeline on the training data
pipeline.fit(X_train, y_train)

# Evaluate the model
train_score = pipeline.score(X_train, y_train)
test_score = pipeline.score(X_test, y_test)

print("Training score:", train_score)
print("Test score:", test_score)

```

```python
# Make prediction on the first instance of the training set
first_instance_prediction = pipeline.predict(X_train[:1])

print("Prediction on the first instance of training set:", first_instance_prediction)
```

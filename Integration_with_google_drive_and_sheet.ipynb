# Pydrive

````python
from pydrive2.auth import GoogleAuth    # GoogleAuth is used to authenticate with Google Drive using OAuth2.
from pydrive2.drive import GoogleDrive  # provides an interface for interacting with Google Drive files and folders.
from google.colab import auth   #The auth module provides functions for authenticating with Google services within Colab notebooks.
from oauth2client.client import GoogleCredentials # GoogleCredentials represents a user's OAuth2 credentials for accessing Google APIs.
````

````python
auth.authenticate_user()

gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
mypydrive = GoogleDrive(gauth)
````

# Auto-iterate through all files that matches this query
```python
file_list = mypydrive.ListFile({'q': "'root' in parents and trashed=false"}).GetList()

  print('title: %s, id: %s' % (file1['title'], file1['id']))
```

Here's an example of how you can use the mypydrive object to list the files in your Google Drive root folder:
```python
files = mypydrive.ListFile({'q': "'root' in parents and trashed=false"}).GetList()
for file in files:
  print('title: %s, id: %s' % (file['title'], file['id']))
```

#Lambda function

```python
selector = lambda df: df['A'] > 0.5
```

```python
df_cut = df.loc[selector]
df_cut.shape
```

```python
selector = lambda df: (df['A'] > 0.5)&(df['B'] < 0.2)
# we can also use 'or'(|) or 'not'(~) conditions
```

```python
df_2 = df.loc[selector]
df_2.shape
```
